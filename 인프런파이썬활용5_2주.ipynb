## 7.데이터 선택 (loc)
   이름을 이용하여 원하는 row에서 원하는 col 선택

```python
import pandas as pd
data = {
    '이름' : ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교' : ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고', '능남고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}
df=pd.DataFrame(data,index=['1번','2번','3번','4번','5번','6번','7번','8번'])
```

```python
df.loc['1번'] #index 1번에 해당하는 전체 데이터
```

```python
df.loc['5번'] #index 5번에 해당하는 전체 데이터
```

```python
df.loc['1번','국어'] #index 1번에 해당하는 국어 데이터
```

```python
df.loc[['1번','2번'],'영어']  #index 1,2번에 해당하는 영어 데이터
```

```python
df.loc[['1번','2번'],['영어','국어']]  #index 1,2번에 해당하는 국어,영어 데이터
```

```python
df.loc['1번':'5번','국어':'사회'] #1번~5번까지 국어부터 사회까지 데이
```

##8.데이터 선택(iloc)
위치를 이용하여 원하는 row에서 원하는 col 선택
i(=integer)

```python
df.iloc[4] #4번째 위치의 데이터
```

```python
df.iloc[0:5] #0부터 4번째까지 데이터 (5포함되지 않음)
```

```python
df.iloc[0,1] #0번쨰 위치의 1번째 (학교) 데이터
```

```python
df.iloc[4,2] #5번 학생의 키 데이터 (4번쨰 위치의 2번쨰 데이터)
```

```python
df.iloc[[0,1],2]  #0,1번째 위치의 학생의 2번째 데이터(키)
```

```python
df.iloc[[0,1],[3,4]] #0,1번째 위치 학생의 3,4번째 데이터 (국어,영어)
```

```python
df.iloc[0:5,3:8] #0~4번째 위치의 학생 중에서 3~7번째 데이터 (국어: 사회)
```

##9.데이터 선택(조건)
조건에 해당하는 데이터 선택

```python
df['키'] >= 185 #학생등릐 키가 185이상인지 여부를 True/False
```

```python
filt=(df['키']>=185)
df[filt]
```

```python
df[-filt]  #filt를 역으로 적용 = 185미만인 사람 출
```

```python
df[df['키']>=185]
```

```python
df.loc[df['키']>=185,'수학'] #키가 185이상인 학생들의 수학 데이터
```

```python
df.loc[df['키']>=185,['이름','수학','과학']] #키가 185이상인 학생들의 이름,수학,과학 데이터
```

###다양한 조건

```python
# & 그리고
df.loc[(df['키']>=185) & (df['학교']=='북산고')] #키가 185이상이고 북산고인 학생 데이
```

```python
# | 또는
df.loc[(df['키']<170) | (df['키']>200)] #키가 170보다 작거나 200보다 큰 학생 데이
```

###데이터 제외하기 dropna

```python
import pandas as pd
data = {
    '이름' : ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교' : ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고', '능남고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}
df=pd.DataFrame(data,index=['1번','2번','3번','4번','5번','6번','7번','8번'])
```

```python
df.dropna() #전체 데이터 중에서 NaN을 포함하는 데이터 삭제
```

```python
df
```

axis = index or column

how = any or all

```python
df.dropna(axis='index',how='any')
```

```python
df.dropna(axis='columns') #NaN이 하나라도 있는 column 삭제
```

```python
df['학교'] = np.nan
df
```

```python
df.dropna(axis='columns',how='all') #데이터 전체가 NaN인 경우에만 column 삭
```
